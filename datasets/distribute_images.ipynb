```python
set_files_path = "datasets/pascal/ImageSets/Segmentation/val.txt"
origin = "datasets/pascal/SegmentationClass/"
origin_extension = ".png"
destination_dir = "datasets/pascal/Segmentation_Output/validation"
```

```python
data = pd.read_csv(set_files_path, header = None)
print(f"moving files to {destination_dir}")
for i, row in enumerate(data.iterrows()):    
    file_name = row[1][0]
    file_path = origin + file_name + origin_extension
    print(f"i = {i+1} / {len(data)}: moving file {file_path}")    
    shutil.copy(file_path, destination_dir)
    
```
